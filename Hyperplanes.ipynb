**What are Hyperplanes?**

A hyperplane is a flat, decision boundary that divides a data space into regions, each associated with a class label. In simple terms, it’s like a line in a two-dimensional space that separates data points of different classes. The concept of a hyperplane extends to higher dimensions: like in three dimensions — it’s a flat plane, and in higher than 3 dimensions — it’s a multi-dimensional space divider.

The dimension of the hyperplane depends upon the number of features. If the number of input features is 2, then the hyperplane is just a line. If the number of input features is 3, then the hyperplane becomes a two-dimensional plane. It becomes difficult to imagine when the number of features exceeds 3. The dimension of the hyperplane is directly proportional to the number of features. As features get increased then the numbers of dimension for ML model also increases and in higher dimension visualization is almost impossible so we need hyperplane which can be easily extended to n dimensions

**Implementation of SVM with 2-Dimensional features:**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

# Generate synthetic 2D data using sklearn's make_blobs
X, y = datasets.make_blobs(n_samples=100, centers=2, random_state=42, cluster_std=1.0)

# Initialize an SVM model with a linear kernel
svm_model = SVC(kernel='linear', C=1.0)
svm_model.fit(X, y)

# Plotting the data points, hyperplane, and margins
plt.figure(figsize=(10, 6))

# Scatter plot of data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', s=50, edgecolors='k', label="Data Points")

# Retrieve model coefficients (weights) and intercept
w = svm_model.coef_[0]
b = svm_model.intercept_[0]

# Calculate the hyperplane line
x_values = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
y_values = -(w[0] * x_values + b) / w[1]
plt.plot(x_values, y_values, 'k-', label="Hyperplane (Decision Boundary)")

# Calculate the margin boundaries (support vectors)
margin = 1 / np.sqrt(np.sum(w ** 2))
y_margin_up = y_values + margin
y_margin_down = y_values - margin

plt.plot(x_values, y_margin_up, 'k--', label="Margin")
plt.plot(x_values, y_margin_down, 'k--')

# support vectors
support_vectors = svm_model.support_vectors_
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], s=100, facecolors='none', edgecolors='k', linewidths=1.5, label="Support Vectors")

# Set labels and title
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("SVM with Hyperplane and Margins in 2-Dimensions")
plt.legend()
plt.show()
```

**Implementation of SVM with 3-Dimensional features:**

```python
from mpl_toolkits.mplot3d import Axes3D

# Generate synthetic 3D data using sklearn's make_blobs
X, y = datasets.make_blobs(n_samples=100, centers=2, n_features=3, random_state=42, cluster_std=1.0)

# Initialize an SVM model with a linear kernel
svm_model = SVC(kernel='linear', C=1.0)
svm_model.fit(X, y)

# Plotting the data points and hyperplane in 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of data points
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='bwr', s=50, edgecolors='k', label="Data Points")

# Get the coefficients of the hyperplane
w = svm_model.coef_[0]
b = svm_model.intercept_[0]

# Create a mesh grid for the hyperplane
x_vals = np.linspace(X[:, 0].min(), X[:, 0].max(), 10)
y_vals = np.linspace(X[:, 1].min(), X[:, 1].max(), 10)
X_vals, Y_vals = np.meshgrid(x_vals, y_vals)
Z_vals = -(w[0] * X_vals + w[1] * Y_vals + b) / w[2]

# Plot the hyperplane
ax.plot_surface(X_vals, Y_vals, Z_vals, color='k', alpha=0.3)

# Highlight support vectors
support_vectors = svm_model.support_vectors_
ax.scatter(support_vectors[:, 0], support_vectors[:, 1], support_vectors[:, 2], s=100, facecolors='none', edgecolors='k', linewidths=1.5, label="Support Vectors")

# Set labels and title
ax.set_xlabel("Feature 1")
ax.set_ylabel("Feature 2")
ax.set_zlabel("Feature 3")
ax.set_title("3D SVM with Hyperplane and Support Vectors")
ax.legend()
plt.show()
```
